```python
#Thêm các thư viện cần thiết
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Đọc dữ liệu data
data = pd.read_csv("/content/drive/MyDrive/PTDL&HS/patient_heart_rate.csv")
data.head(5)
```

##Cau 1

```python
#Tạo mảng col_name chứa tên giá trị các cột
#đọc data và gán tên các cột theo từng vị trí trong mảng col_name
col_name = ["Id",'Name','Age',"Weight",'m0006','m0612','m1218','f0006','f0612','f1218']
data = pd.read_csv("/content/drive/MyDrive/PTDL&HS/patient_heart_rate.csv",names= col_name)
print(data)
```

##Cau 2

```python
#Tạo 2 cột fristname và lastname từ cột name sau khi đã tách các chuỗi bằng dấu cách và ứng với mỗi chuỗi đã tách sẽ thêm vào từng cột tương ứng
#Xóa cột Name
data[["Firstname","Lastname"]] = data["Name"].str.split(expand=True)
data = data.drop("Name",axis=1)
print(data)
```

##Cau 3

```python
# Chuẩn hóa dữ liệu trên cột Weight thành cùng 1 kiểu là kgs.
weight = data["Weight"]
for i in range(0,len(weight)):
  x = str(weight[i])
  if "lbs" in x[-3:]:
    x = x[:-3:]
    float_x = float(x)
    y = int(float_x/2.2)
    y = str(y)+'kgs'
    weight[i] = y
```

```python
#xóa hết tất cả các cột trùng nhau
data.dropna(how="all",inplace= True)
data
```

##Cau 5

```python
#Xóa các dòng dữ liệu bị trùng lặp trên 4 cột "Firstname","Lastname","Age","Weight"
data.drop_duplicates(subset=["Firstname","Lastname","Age","Weight"])
data
```

##Cau 6

```python
# Chuẩn hóa dữ liệu của các chữ không phải là ASCII.
data.Firstname.replace({r'[^\x00-\x7F]+':''},regex=True,inplace=True)
data.Lastname.replace({r'[^\x00-\x7F]+':''},regex=True,inplace=True)
data
```

##Cau 7

```python
# tìm giá trị null trên 2 cột Age và Weight
index_null = data[data["Age"].isna() & data["Weight"].isna()].index
index_null
```

```python
#Xóa dữ liệu trên các chỉ mục đó
data.drop(index_null,inplace = True)
```

```python
#fill giá trị trung bình trên cột age
data["Age"].fillna(data["Age"].mean(),inplace=True)
data
```

##Cau 8

```python
#Giữ lại các cột "Id",'Age','Weight',"Firstname","Lastname" và chuyển các cột còn lại thành dữ liệu cho cột tạo sex_and_time và PulseRate sẽ thành cột chứa các giá trị của các cột trên
data = pd.melt(data,id_vars=["Id",'Age','Weight',"Firstname","Lastname"],value_name="PulseRate",var_name = "sex_and_time").sort_values(["Id",'Age','Weight',"Firstname","Lastname"])
```

```python
# Tách các thông tin giới tính và ngày sau đó gán chúng vào data frame tạm.

tmp_df= data["sex_and_time"].str.extract("(\D)(\d+)(\d{2})", expand=True)
tmp_df.head(10)
```

```python
tmp_df.columns = ["sex","hour","minute"]
tmp_df.head(10)
```

```python
# Tạo cột Time và chuẩn hóa dữ liệu.
tmp_df["Time"] = tmp_df["hour"]+"-"+tmp_df["minute"]
tmp_df.head(10)
```

```python
data.head(10)
```

```python
#Nhập các dataframe tạm vào data chính
data = pd.concat([data,tmp_df],axis=1)
data.head(10)
```

```python
#Xóa các cột "sex_and_time","hour","minute"

data = data.drop(["sex_and_time","hour","minute"],axis=1)
data = data.dropna()
data.head(10)
```

```python
data['PulseRate']= data['PulseRate'].replace({'-' : '0'}, regex=True)

data
```

```python
# Tính giá trị trung bình
def avgPulserate(df, firstname):
  sumPulserate = 0
  nsumPulserate = 0
  for i, value in enumerate(data['Firstname']):
    if value == firstname:
      try:
        if data['PulseRate'][i] !='0':
          sumPulserate += int(data['PulseRate'].values[i])
          nsumPulserate += 1
      except:
        pass
  if nsumPulserate > 0:
    return str(sumPulserate/nsumPulserate)
  else:
    return 0
```

```python
# Kiểm tra vị trí hợp lệ
def isValidIndex1(index, nrows):
  return index - 1 > 0 & index + 1 < nrows
```

```python
# Kiểm tra vị trí hợp lệ
def isValidIndex2(index, nrows):
  return index + 1 < nrows & index + 2 < nrows
```

```python
# Kiểm tra giá trị không null
def isPulseRateNotNull1(index, df):
  return str(df['PulseRate'].values[index-1])!= '0' and str(df['PulseRate'].values[index+1])!= '0'
```

```python
# Kiểm tra giá trị không null
def isPulseRateNotNull2(index, df):
  return str(df['PulseRate'].values[index+1])!= '0' and str(df['PulseRate'].values[index +2])!= '0'
```

```python
#Kiếm tra giá trị có trong dòng
def isInGroup1(df, index, firstname):
  return df['Firstname'].values[index-1]== firstname and df['Firstname'].values[index +1]== firstname
```

```python
#Kiếm tra giá trị có trong dòng
def isInGroup2(df, index, firstname):
  return df['Firstname'].values[index+1]== firstname and df['Firstname'].values[index +2]== firstname
```

```python
#Update dữ liệu từ cột PulseRate
def updatePulseRatena (index, df):
  nrows = len(df['PulseRate']) -2
  newvalue = ''
  firstname = df['Firstname'].values[index]
  try:
    if isValidIndex1(index,nrows) & isPulseRateNotNull1(index, df) & isInGroup1(df, index, firstname):
      try:
        newvalue = (int(df['PulseRate'].values[index-1]) + int(df['PulseRate'].values[index+1]))/2
      except:
        newvalue = '0'
    elif isValidIndex2(index,nrows) & isPulseRateNotNull2(index, df) & isInGroup2(df, index, firstname):
      try:
        newvalue = (int(df['PulseRate'].values[index-1]) + int(df['PulseRate'].values[index+1]))/2
      except:
        newvalue ='0'
    else:
      newvalue = avgPulserate(df, firstname)
  except:
      newvalue = avgPulserate(df, firstname)
  finally:
    pass
  return newvalue
```

```python
#Tạo mảng tạm lưu các giá trị update sau khi dùng hàm trên
pulserateupdate = []
for index, value in enumerate(data["PulseRate"]):
  if value =='0' or value == '':
    newvalue = updatePulseRatena(index,data)
    pulserateupdate.append(newvalue)
  else:
    pulserateupdate.append(value)
```

```python
#Tạo cột PulseRateUpdate và truyền dữ liệu từ mảng tạm vào cột
data["PulseRateUpdate"] = pulserateupdate
```
