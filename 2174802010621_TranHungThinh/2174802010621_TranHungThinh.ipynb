```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

##Cau 1: Tao DataFrame tu du lieu

```python
data = pd.DataFrame({"Name" : ["Alice","Bob","Charlie","David","Eva","Frank","Grace","Hannah","Ivan","Jack","Kelly","Liam","Mona","Nina","Oscar"],
                    "Age" : [25,30,35,28,22,45,34,31,27,29,33,40,26,32,36],
                    "Salary": [50000,60000,70000,55000,52000,80000,72000,68000,61000,59000,63000,77000,53000,66000,75000]
                    })
```

##Cau 2: Hien thi thong tin DataFrame

```python
data.info()
```

##Cau 3: Loc cac hang trong DataFrame co Age > 28

```python
data_check = data[data["Age"] > 28]
data_check
```

##Cau 4: Tinh gia tri trung binh cot Salary

```python
avg = 0
for i in data["Salary"].values:
    avg += i
ans = avg / len(data["Salary"].values)
print(ans)
```

```python
data_4 = data["Salary"].mean()
data_4
```

##Cau 11: Kiem tra gia tri NaN co trong DataFrame khong

```python
data_checked = data.isnull()
data_checked
```

##Cau 12: Thay the gia tri cua cot Age lon hon 30 bang gia tri trung binh cua cot do

```python
tmp = []
for i in data["Age"].values:
    if i > 30:
        tmp.append(int(data["Age"].mean()))
    else:
        tmp.append(i)
data["Age"] = tmp
data
```

##Cau 13: Chuan hoa cot Age ve khoang gia tri 0 va 1

```python
data_check_13 = (data["Age"] - data["Age"].min()) / (data["Age"].max() - data["Age"].min())
data_check_13
```

##Cau 14: Tao mot cot Age-group phan loai tuoi

```python
tmp = []
for i in data["Age"].values:
  if i < 30: tmp.append("Young")
  elif i >= 30 and i < 65: tmp.append("Middle-aged")
  else: tmp.append("Old")
data["Age_group"] = tmp
data
```

##Cau 15: tinh toan ti le phan tram thay doi (percentage change) cua cot Salary

```python
percent_change = data["Salary"].pct_change()
percent_change
```
