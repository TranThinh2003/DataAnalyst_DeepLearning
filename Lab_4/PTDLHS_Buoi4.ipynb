```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#Câu 1

```python
#Cau 1
df = pd.read_csv("/content/drive/MyDrive/PTDL&HS/Lab_4/titanic_disaster.csv")
df.head(10)
```

#Câu 2

```python
#Cau 2
sns.heatmap(df.isna(),
            yticklabels=False,
            cbar=True,
            cmap='viridis')
```

#Câu 3

```python
#Cau 3
#Tạo các cột firstname và secondname để lưu các giá trị của cột name sau khi tách các từ theo dấu ,
#Sau đó xóa cột name
df[["firstName","secondName"]] = df["Name"].str.split(",",expand=True)
data = df.drop("Name",axis=1)
print(data)
```

#Câu 4

```python
#Cau 4
#Thay đổi các chỗ có giá male thành M và female thành F
df.replace({"male":"M","female":"F"},inplace=True)
df.head(10)
```

```python
#trực quan dữ liệu bằng hàm boxplot
sns.boxplot(x="Pclass",y="Age",data=df)
plt.show()
```

#Câu 5

```python
#Cau 5
#điền các giá trị tại các chỗ có giá trị rỗng thành giá trị trung bình của cột tuổi
df['Age'].fillna(int(round(df['Age'].mean())), inplace= True)
df.head(10)
```

```python
df["Age"]
```

```python
sns.heatmap(df.isnull(),
            yticklabels=False,
            cbar=True,
            cmap='viridis')
```

#Câu 6

```python
#Cau 6
# tạo mảng tạm để lưu kết quả
# qua mỗi giá trị trong cột age xét điều kiện và lưu kết quả trong mảng tạm
# kết thúc vòng lặp gán giá trị của mảng Agegroup là mảng tạm
tmp = []
for i in df ["Age"].values:
  if i <= 12: tmp.append("Kid")
  elif i > 12 and i <= 18: tmp.append("Teen")
  elif i > 18 and i <= 60: tmp.append("Adult")
  else: tmp.append("Older")
df["Age_group"] = tmp
df.head(10)
```

#Câu 7

```python
#Cau 7
#tạo một dataframe tạm và tách xâu theo dấu . thành các xâu con ứng với từng cột
#gán giá trị tại cột đầu trong dataframe tạm vào cột tạo nameprefix trong dataframe gốc
df_tmp = df["secondName"].str.split(".",expand=True)
df["namePrefix"] = df_tmp[0]
df.head(10)
```

#Câu 8

```python
#Cau 8
#tính toán các giá trị tại các hàng tương ứng với các cột đang xét
df["familySize"] = df["SibSp"] + df["Parch"] + 1
df.head(10)
```

#Câu 9

```python
#Cau 9
#tạo mảng tạm lưu kết quả
#qua mỗi giá trị trong cột familySize xét điều kiện và lưu kết quả trong mảng tạm
#gán giá trị cho cột Alone là mảng tạm
tmp = []
for i in df["familySize"].values:
  if i == 1: tmp.append(1)
  else: tmp.append(0)
df["Alone"] = tmp
df.head(10)
```

#Câu 10

```python
#Cau 10
#tạo mảng tạm lưu kết quả
#qua mỗi giá trị trong cột Cabin xét điều kiện và lưu kết quả trong mảng tạm
#gán giá trị cho cột typeCabin là mảng tạm
tmp = []
for i in df["Cabin"].values:
  if i is np.nan: tmp.append("Unknown")
  else: tmp.append(i[0])
df["typeCabin"] = tmp
df.head(10)
```

#Câu 15

```python
# Trực quan hóa dữ liệu với đồ thị catplot với tham số x là familySize sẽ vẽ trên trục x và y là Survived sẽ vẽ trên trục y
# kind = point: tạo ra một biểu đồ phân tán với các điểm đại diện cho từng điểm dữ liệu.
sns.catplot(x ='familySize', y ='Survived', data = df, kind ='point')
plt.title("Survived by Family Size")
plt.xlabel("Family Size")
plt.ylabel("Survived")
plt.show()
```

```python
# Trực quan hóa dữ liệu với đồ thị catplot với tham số x là Fare sẽ vẽ trên trục x và y là Survived sẽ vẽ trên trục y
# kind = point: tạo ra một biểu đồ phân tán với các điểm đại diện cho từng điểm dữ liệu.
sns.catplot(x ='Fare', y ='Survived', data = df, kind ='point')
plt.title("Survived by Fare")
plt.xlabel("Fare")
plt.ylabel("Survived")
plt.show()
```

```python
# Trực quan hóa dữ liệu với đồ thị catplot với tham số x là Pclass sẽ vẽ trên trục x và y là Survived sẽ vẽ trên trục y
# kind = "point": tạo ra một biểu đồ phân tán với các điểm đại diện cho từng điểm dữ liệu.
# hue = "sex": màu săc khác nhau đại diện cho từng giới tính khác nhau
# col='Embarked': tạo ra một bố cục lưới khía cạnh và sẽ được chia thành các ô phụ dựa trên biến phân loại "Embarked".
# Mỗi ô phụ sẽ hiển thị mối quan hệ giữa "Pclass" và "Survived" đối với những hành khách khởi hành từ một cảng cụ thể.
sns.catplot(x='Pclass', y='Survived', hue='Sex', col='Embarked', data=df,kind='point')
plt.show()
```
