```python
#
def Q1(g):
  return np.percentile(g, 25)
def Q2(g):
  return np.percentile(g, 50)
def Q3(g):
  return np.percentile(g, 75)
def std(g):
  return np.std(g)
def mean(g):
  return sum(g) / len(g)
```

```python
#Cau 3
#là bảng xem tổng quan bộ dữ liệu
pd.pivot_table(data=data, index='KT', values="DH1", aggfunc=['sum','mean','median','min','max',std,Q1,Q2,Q3])
```

```python
#Cau 4
pd.pivot_table(data=data, index=['KT',"KV"], values="DH1", aggfunc=['sum','mean','median','min','max',std,Q1,Q2,Q3])
```

```python
#Cau 2
# Tạo cột mới là phanlopt1 xong sau đó truyen tu T1 sang theo điều kiện của hàm
def phanlopt1(t1):
  if t1 < 5: return 'k'
  elif t1 < 7: return 'tb'
  elif t1 < 8: return 'kha'
  else: return 'g'
data['phanlopt1'] = data["T1"].apply(phanlopt1)
```

```python
#Cau 3
#chỉ điểm đến chỗ data thỏa điều kiện và gán xếp lực
data.loc[(data['T1'] < 5),'phanlop1'] = 'kem'
data.loc[(data['T1'] >= 5) & data['T1'] < 7,'phanlop1'] = 'trungbinh'
data.loc[(data['T1'] >= 7) & data['T1'] < 8,'phanlop1'] = 'kha'
data.loc[(data['T1'] >= 8),'phanlop1'] = 'gioi'
```

```python
data.head(10)
```

```python
data.groupby('phanlopt1')['phanlopt1'].count().plot()
```

```python
#Cau 4

data_phanlop1 = data.groupby(["T1",'phanlopt1'])[["T1","phanlopt1"]].count()
unstackphanlop1 = data_phanlop1.unstack()
unstackphanlop1.head(10)
```

```python
#dùng hàm plot để visualize
unstackphanlop1["phanlopt1"].plot(ylabel='Count')
```
