# Câu 4

```python
# Thay thế các dữ liệu còn thiếu bằng 0
read.fillna(0,inplace = True)
read.tail(10)
```

#Câu 5

```python
# Thay thế các dữ liệu bị thiếu thành các giá trị trung bình
# read.T1.mean() tính toán giá trị trung bình của cột T1
read["T1"] = read.T1.fillna(read.T1.mean(),inplace= True)
read.head(10)
```

#Câu 6

```python
# Hàm để điền các giá trị còn thiếu trong dữ liệu
def fill(col_name):
  return read[col_name].fillna(read[col_name].mean(),inplace = False)
```

```python
# Đối với từng phần tử trong vòng lặp for i sẽ cộng lần lượt với từng phần tử trong vòng for j tương ứng giống với tên các cột điểm trong data
# Dùng hàm fill để bổ sung các dữ liệu còn thiếu với từng vị trí ở trong cột
subject = ["T","L","H","S","V","X","D","N"]
classes = ["1","2","6"]
for i in classes:
  for j in subject:
    fill(j+i)
read.tail(10)
```

#Câu 7

```python
# Vòng for thực hiện thao tác tạo biến TBM1 và tính toán. Sau đó lặp lại thao tác trên tạo biến TBM2 và tính toán. TBM3 tương tự
for i in ["1",'2','3']:
  read['TBM' + i] = (read["T"+i]*2 + read["L"+i] + read["H"+i] + read["S"+i]+read["V" + i]*2 +read["X" + i] +read["D" + i] +read["N" + i] )/10
read.tail(10)
```

#Câu 8

```python
# Qua mỗi vòng lặp for i, biến data sẽ lưu mảng gồm các giá trị tại cột tương ứng với từng phần tử trong mảng ["1","2","3"]. (read["TBM" + i].values)
# Qua mỗi vòng lặp for j, biến tmp sẽ lưu các giá trị kí tự ("Y", "TB", "K",...) thông qua các điều kiện đối với từng phần tử trong data
# Sau khi lặp xong vòng for j thì tạo cột và đổ dữ liệu vào cột đã tạo
# Lặp lại bước 1 cho đến khi hết vòng lặp i
for i in ["1","2","3"]:
  data = read["TBM" + i].values
  tmp = []
  for j in data:
    if j < 5.0: tmp.append("Y")
    elif j >= 5.0 and j <= 6.5: tmp.append("TB")
    elif j >= 6.5 and j < 8.0: tmp.append("K")
    elif j >= 8.0 and j < 9.0: tmp.append("G")
    else: tmp.append("XS")
  read["XL" + i] = tmp
```

#Câu 9

```python
#Qua mỗi vòng lặp for i, biến data sẽ lưu mảng gồm các giá trị tại cột tương ứng với từng phần tử trong mảng ["1","2","3"]. (read["TBM" + i].values)
#Qua mỗi vòng lặp for j, biến tmp sẽ lưu kết quả của từng giá trị tại từng cột
#Truyền dữ liệu từ tmp vào từng cột sau khi lặp xong vòng for j và tiếp tục lặp lại b1 cho đến khi hết vòng lặp for i
for i in ["1", "2", "3"]:
    data = read["TBM"+i].values
    tmp = []
    for j in data:
        tmp.append((j*4)/10)
    read["US_TBM"+i] = tmp
```

#Câu 10

```python
# Tạo các biến DH1, DH2, DH3, KT để lưu các giá trị tại các cột tương ứng
# Vòng lặp for i sẽ lặp đến 100 lần (len(DH1) = 100) đối với mỗi lần lặp check điều kiện phần tử tại vị trí thứ i thuộc khối thi nào (KT)
# Đối với mỗi khoa thi tương ứng thì sẽ lưu các giá trị 1 và 0 và mảng tạm tmp vào các vị trí tự i tương ứng
# Cuối cùng tạo cột KQXT và truyền dữ liệu từ mảng tmp sang
DH1 = read["DH1"].values
DH2 = read["DH2"].values
DH3 = read["DH3"].values
KT = read["KT"].values
tmp = []
A_major = ['A','A1']
for i in range(len(DH1)):
  if KT[i] in A_major:
    if (DH1[i]*2 + DH2[i] + DH3[i])/4 >= 5.0: tmp.append(1)
    else: tmp.append(0)
  elif KT[i] == "B":
    if (DH1[i] + DH2[i]*2 + DH3[i])/4 >= 5.0: tmp.append(1)
    else: tmp.append(0)
  else:
    if (DH1[i] + DH2[i] + DH3[i])/3 >= 5.0: tmp.append(1)
    else: tmp.append(0)
read["KQXT"] = tmp
```
